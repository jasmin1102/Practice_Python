```python
import openpyxl

def excel_create_func(filename, sheet_title, list_data):
    excel_file = openpyxl.Workbook()
    excel_sheet = excel_file.active
    excel_sheet.title = sheet_title
    excel_sheet.column_dimensions['A'].width = 200
    excel_sheet.column_dimensions['B'].width = 30
    excel_sheet.append(['상품명','상품가격'])
    
    for eachlist in list_data:
        excel_sheet.append(eachlist)
        excel_file.save(filename)
        excel_file.close()

goods_list = list()

import requests
from bs4 import BeautifulSoup

res = requests.get('https://search.shopping.naver.com/search/all?query=%EC%88%98%EC%A0%9C+%EA%B7%B8%EB%9E%98%EB%86%80%EB%9D%BC&cat_id=&frm=NVSHATC')
soup = BeautifulSoup(res.content,'html.parser')
items = soup.select('div.basicList_info_area__17Xyo')

for goods in items:
    goods_title = goods.select_one('div.basicList_title__3P9Q7')
    goods_price = goods.select_one('div.basicList_price_area__1UXXR')
    goods_info = [goods_title.get_text().strip(), goods_price.get_text().strip()]
    goods_list.append(goods_info)
```

```python
excel_create_func('수제 그래놀라 엑셀(크롤링 기초 버전).xlsx', '순위 리스트', goods_list)
```
