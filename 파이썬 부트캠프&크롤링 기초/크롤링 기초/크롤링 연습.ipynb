```python
from bs4 import BeautifulSoup
html = """
        <html> \
          <body> \
              <h1 id='title'>[1]크롤링이란?</h1> \
              <p class='cssstyle'>웹페이지에서 필요한 데이터를 추출하는 것</p> \
            <p id='body' align='center'>파이썬을 중심으로 다양한 웹크롤링 기술 발달</p> \
          </body> \
        </html> 
""" 
# 하나의 문자열이 여러 줄에 걸쳐있을 때, 각 줄 마지막에 역슬레쉬'\'쓰거나, """ 큰따옴표 3개 쓰기

soup = BeautifulSoup(html, "html.parser")
# ↑해당 코드의 경우. html 변수에 html문자열을 그대로 넣었기에 html.content와 같은 코드는 필요 없음

data = soup.find('p',attrs = {'id':'body', 'align':'center'})
print(data.string) # = print(data.get_text())

print('\n')
 
data2 = soup.find_all('p')
for item in data2:
    print(item.string)
```

#### 웹페이지에서 필요한 데이터를 추출하기 위해 태그에 담긴 속성까지도 선택하는 방법
1. data = soup.find('p', class_ = 'cssstyle') : ('태그명', 속성명 = '속성값')
- p라는 태그 class = 'csstyle'의 속성
- '_ (언더바)'를 붙이는 이유: 객체지향에서 나오는 'class'와 구분하기 위해 불가피


2. data = soup.find('p', 'csstyle') : ('태그명', '속성값)


3. data = soup.find('p', attrs = {'align':'center'})
- attrs = {'속성이름1':'속성값1', '속성이름2':'속성값2', ...}

4. data = soup.find('태그', id = '속성')
- 'id'라는 속성은 특정 데이터에만 붙여지는 유니크한 이름을 붙이는 것. 따라서 별도의 문법 없이 'id' = '속성값'만 넣어도 ok
- class나 id는 흔히 쓰이는 속성이기에 굳이 별도의 문법없이 'id/class' = '속성값' 만 입력해도 됨

#### find_all('????')
: 'find_all('????')' = ????라는 특정 태그(및 속성)에 해당하는 데이터를 html파일에서 모두 가져오는 함수.

: 이때, find_all()로 찾은 데이터는 여러 개 이므로 '리스트' 데이터 구조를 사용함
- for문 사용하기

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://v.media.daum.net/v/20170615203441266')
soup = BeautifulSoup(res.content, 'html.parser')

mydata = soup.find('div', 'layer_util layer_summary') 
#'class'는 이미 많이 쓰이는 속성이므로 속성값만 넣어도 됨
# class 이름이 여러 개일 경우, 해당 class 이름 전부 넣어주면 됨
for item in mydata:
    print(item.string)
```
