```python
import requests
from bs4 import BeautifulSoup

site_list = ['https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000149','https://search.shopping.naver.com/best100v2/detail.nhn?catId=50000000']

for site in site_list:
    res = requests.get(site) # 크롤링하려는 웹페이지 개수가 여러 개일 경우, 웹사이트 주소를 리스트에 넣고 requests.get(리스트 명)으로 불러오기!
    soup = BeautifulSoup(res.content, 'html.parser')
    data = soup.select('#productListArea > ul > li > p > a')
    print(site)
    for index, item in enumerate(data):
        print(1+index,"위",item.get_text())
```

```python
import requests
from bs4 import BeautifulSoup

res = requests.get('https://finance.naver.com/')
soup = BeautifulSoup(res.content,'html.parser')

items = soup.select('#container > div.aside > div > div.aside_area.aside_popular > table > tbody > tr')

for item in items:
    print("종목명:",item.find('a').get_text(),"/","현재가:",item.select_one('td').get_text(),"/","전일대비:",item.select_one('span').get_text())
```

```python
#copyelement 이용하여 추출하고자하는 부분의 데이터를 태그별로 나누기 & 

import requests
from bs4 import BeautifulSoup

res = requests.get('https://finance.naver.com/sise/')
soup = BeautifulSoup(res.content, 'html.parser')
items = soup.select("div.rgt > ul.lst_major > li")

for item in items:
    print("지수명:",item.find('a').get_text(), "현재지수:",item.find('span').get_text(), "등락:",item.find('em').get_text())
    #<a>~~~</a> 'a'태그끼리 묶인 부분만 추출. <span>~~~</span> 'span'태그끼리 묶인 부분만 추출
```
