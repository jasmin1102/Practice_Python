```python

import requests
from bs4 import BeautifulSoup

res = requests.get("http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G06")
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select('#gBestWrap > div > div:nth-child(5) > div:nth-child(3) > ul > li')

for index, each in enumerate(data):
    title = each.select_one('li > a')
    price = each.select_one('div.item_price > div.s-price > strong > span > span')
    
   #크롤링한 데이터에서 또다른 데이터 크롤링하기 
    res_info = requests.get(title['href'])  # 딕셔너리처럼~~ '변수명['속성명']'
    soup_info = BeautifulSoup(res_info.content, 'html.parser')
    provider_info = soup_info.select_one('div.item-topinfo_headline > p > span.text__seller > a')
    print(str(index+1),title.get_text(), price.get_text(), provider_info.get_text())
```

```python
import openpyxl

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.title = '상품정보리스트'
excel_sheet.column_dimensions['B'] = 150
excel_sheet.column_dimensions['C'] = 20
excel_sheet.column_dimensions['D'] = 150
excel_sheet.column_dimensions['E'] = 30
excel_sheet.append(['순위','상품명','판매가격','상품상세링크','판매자'])
    
        
import requests
from bs4 import BeautifulSoup

res = requests.get("http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G06")
soup = BeautifulSoup(res.content, 'html.parser')
data = soup.select('#gBestWrap > div > div:nth-child(5) > div:nth-child(3) > ul > li')

for index, each in enumerate(data):
    title = each.select_one('li > a')
    price = each.select_one('div.item_price > div.s-price > strong > span > span')
    
   #크롤링한 데이터에서 또다른 데이터 크롤링하기 
    res_info = requests.get(title['href'])  # 딕셔너리처럼~~ '변수명['속성명']'
    soup_info = BeautifulSoup(res_info.content, 'html.parser')
    provider_info = soup_info.select_one('div.item-topinfo_headline > p > span.text__seller > a')
    total_info = [str(index+1),title.get_text(), price.get_text(),title['href'], provider_info.get_text()]
    excel_sheet.append(total_info)
    excel_sheet.cell(row=index+2,column=4).hyperlink =title['href']
    
cell_A1 = excel_sheet['A1']
cell_A1.alignment = openpyxl.styles.Alignment(horizontal='center') #셀A1에 있는 '랭킹' 문자열을 가운데에 배치하고자 할 때
cell_A1.font = openpyxl.styles.Font(color="F8BBD0")

cell_B1 = excel_sheet['B1']
cell_B1.alignment = openpyxl.styles.Alignment(horizontal='center') #셀A1에 있는 '랭킹' 문자열을 가운데에 배치하고자 할 때
cell_B1.font = openpyxl.styles.Font(color="F8BBD0")

cell_C1 = excel_sheet['C1']
cell_C1.alignment = openpyxl.styles.Alignment(horizontal='center') #셀A1에 있는 '랭킹' 문자열을 가운데에 배치하고자 할 때
cell_C1.font = openpyxl.styles.Font(color="F8BBD0")

cell_D1 = excel_sheet['D1']
cell_D1.alignment = openpyxl.styles.Alignment(horizontal='center') #셀A1에 있는 '랭킹' 문자열을 가운데에 배치하고자 할 때
cell_D1.font = openpyxl.styles.Font(color="F8BBD0")

cell_E1 = excel_sheet['E1']
cell_E1.alignment = openpyxl.styles.Alignment(horizontal='center') #셀A1에 있는 '랭킹' 문자열을 가운데에 배치하고자 할 때
cell_E1.font = openpyxl.styles.Font(color="F8BBD0")

excel_file.save('crawling_crawling실습본.xlsx')
excel_file.close()
```
