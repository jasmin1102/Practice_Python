```python
import requests
from bs4 import BeautifulSoup

res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G01&subGroupCode=S161&largeCategoryCode=100000103&mediumCategoryCode=200002671')
soup = BeautifulSoup(res.content, 'html.parser')
bestlists = soup.select('div.best-list')
bestitems = bestlists[1]
products = bestitems.select('ul > li')

for index, product in enumerate(products):
    title = product.select_one('a.itemname')
    price = product.select_one('div.s-price > strong')
    print(str(index+1),title.get_text(),price.get_text(),title['href'])
```

```python
import requests
from bs4 import BeautifulSoup

import re
link_re = re.compile('^http://')

res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G01&subGroupCode=S161&largeCategoryCode=100000103&mediumCategoryCode=200002671')
soup = BeautifulSoup(res.content,'html.parser')
bestlists = soup.select('div.best-list')
bestitems = bestlists[1]
products = bestitems.select('ul > li')

for index, product in enumerate(products):
    title = product.select_one('a.itemname')
    price = product.select_one('div.s-price > strong')
    if link_re.match(title['href']): #match는 search와 달리 일부만 성립해도 된다!
        res_info = requests.get(title['href'])
        soup_info = BeautifulSoup(res_info.content, 'html.parser')
        provider_info = soup_info.select_one('div.item-topinfo_headline > p > span.text__seller > a')
        
        print(title.get_text(), price.get_text(), title['href'],provider_info.get_text())
```

```python
import requests, openpyxl
from bs4 import BeautifulSoup

import re
link_re = re.compile('^http://')

excel_file = openpyxl.Workbook()
excel_sheet = excel_file.active
excel_sheet.append(['랭킹','상품명','판매가격','상품상세링크','판매자정보'])
excel_sheet.column_dimensions['B'].width = 80
excel_sheet.column_dimensions['C'].width = 20
excel_sheet.column_dimensions['D'].width = 80
excel_sheet.column_dimensions['E'].width = 20

res = requests.get('http://corners.gmarket.co.kr/Bestsellers?viewType=G&groupCode=G01&subGroupCode=S161&largeCategoryCode=100000103&mediumCategoryCode=200002671')
soup = BeautifulSoup(res.content, 'html.parser')
bestlists = soup.select('div.best-list')
bestitems = bestlists[1]
products = bestitems.select('ul > li')

for index,product in enumerate(products):
    title = product.select_one('a.itemname')
    price = product.select_one('div.s-price > strong')
    if link_re.match(title['href']):
        res_1 = requests.get(title['href']) #개별 상품 페이지 접속 주소(판매자 정보는 개별 상품 페이지에 있으므로, 개별상품페이지 접속 주소가 반드시 필요함)
        soup_1 = BeautifulSoup(res_1.content, 'html.parser')
        provider_info = soup_1.select_one('div.item-topinfo_headline > p > span.text__seller > a')
        
        excel_sheet.append([index+1, title.get_text(), price.get_text(), title['href'], provider_info.get_text()])
        excel_sheet.cell(row = index+2, column = 4).hyperlink = title['href']
        
cell_A1 = excel_sheet['A1']
cell_A1.alignment = openpyxl.styles.Alignment(horizontal = 'center')
cell_A1.font = openpyxl.styles.Font(color='64FFDA')

cell_B1 = excel_sheet['B1']
cell_B1.alignment = openpyxl.styles.Alignment(horizontal = 'center')
cell_B1.font = openpyxl.styles.Font(color='64FFDA')

cell_C1 = excel_sheet['C1']
cell_C1.alignment = openpyxl.styles.Alignment(horizontal = 'center')
cell_C1.font = openpyxl.styles.Font(color='64FFDA')

cell_D1 = excel_sheet['D1']
cell_D1.alignment = openpyxl.styles.Alignment(horizontal = 'center')
cell_D1.font = openpyxl.styles.Font(color='64FFDA')

cell_E1 = excel_sheet['E1']
cell_E1.alignment = openpyxl.styles.Alignment(horizontal = 'center')
cell_E1.font = openpyxl.styles.Font(color='64FFDA')

excel_file.save('crawling_practice.xlsx')
excel_file.close()
```
